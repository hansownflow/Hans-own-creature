```python
import random
import time
import matplotlib.pyplot as plt
```

```python
def insertion_sort(arr):  # mendefinisikan sebuah fungsi
    for i in range(1, len(arr)):  # melakukan iterasi sampai akhir list
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:  # melakukan iterasi mundur
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
```

```python
# List untuk menyimpan waktu proses sorting
sorting_times = []

# List untuk menyimpan jumlah angka yang diurutkan
number_of_elements = []
```

```python
# Loop untuk menguji algoritma sorting dengan berbagai ukuran array
# Menggunakan range dari 10 hingga 160 dengan selang 20
for n in range(10, 161, 20):
    arr = random.sample(range(1, 1000), n)  # Input data range 1-1000
    print(f"\nUji dengan {n} elemen:")
    print("Angka sebelum Diurutkan (5 pertama): ")
    print(arr[:5])  # Hanya menampilkan 5 elemen pertama agar output tidak terlalu panjang

    start_time = time.time()
    insertion_sort(arr)
    end_time = time.time()

    print("Angka setelah Diurutkan (5 pertama): ")
    print(arr[:5])
    sorting_time = end_time - start_time
    print(f"Waktu eksekusi: {sorting_time:.6f} detik")

    sorting_times.append(sorting_time)
    number_of_elements.append(n)
```

```python
# Membuat plot menggunakan matplotlib
plt.figure(figsize=(10, 6))
plt.plot(number_of_elements, sorting_times, marker='o', linestyle='-', color='blue')
plt.title('Running Time vs Jumlah Angka yang Diurutkan (Insertion Sort)')
plt.xlabel('Jumlah Angka')
plt.ylabel('Waktu Proses Sorting (detik)')
plt.grid(True)
plt.xticks(number_of_elements)  # Menampilkan semua nilai n pada sumbu x
plt.tight_layout()
plt.show()

print("\nHasil Pengukuran:")
for i in range(len(number_of_elements)):
    print(f"Jumlah elemen: {number_of_elements[i]}, Waktu: {sorting_times[i]:.6f} detik")
```
